```python
# Realizar una función graficar_pol(pol) que tome como parametro un polinomio y lo gráfique
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargaré acá sets de polinomios para luego probar
pol_uno = [1, 5, 1] # x^2+5x+1
pol_dos = [1, 0, 0, 1, 5] # x^4 + x + 5
pol_tres = [1, 1, 0, 9] # x^3 + x^2 + 9
pol_cuatro = [1, 0, 0, 0] # x^3 , si no hay termino independiente pondrémos 0
pol_cinco = [1, 0, 1] # x^2 + 1
```

```python
# Función para gráficar polinomios

def graficar_pol(pol):
    # Rango de x:
    x = np.linspace(0, 20, 20)


    # Imagen del pol con x:
    y = np.polyval(pol, x)

    # Gráfico del polinomio:
    plt.plot(x, y, label=f"Polinomio: {escribir_polinomio(pol)}")

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Gráfico del polinomio")


    # Lo mostramos: 
    plt.grid()
    plt.legend()
    plt.show()

# Función reutilizada para reescribir prolijamente el polinomio:

def escribir_polinomio(poli):
    grado = len(poli) - 1
    polinomio = ""

    for i, coef in enumerate(poli):
        exp = grado - i
        if coef != 0:
            # Formatear el término
            if exp == 0:
                termino = f"{round(coef)}"
            elif exp == 1:
                termino = f"{round(coef)}x"
            else:
                termino = f"{round(coef)}x^{exp}"

            # Concatenar el término al polinomio
            if polinomio:
                # Manejar el signo
                if coef > 0:
                    polinomio += " + " + termino
                else:
                    polinomio += " - " + termino.lstrip('-')
            else:
                polinomio += termino

    return polinomio if polinomio else "0"
```

```python
graficar_pol(pol_uno)
graficar_pol(pol_dos)
graficar_pol(pol_tres)
graficar_pol(pol_cuatro)
graficar_pol(pol_cinco)
```

```python
# EXTRA: MODIFICARÁ EL RANGO PARA HACER UNA CUADRATICA:

def graficar_cuadratica(pol):
    # Rango de x:
    x = np.linspace(-20, 20, 20)


    # Imagen del pol con x:
    y = np.polyval(pol, x)

    # Gráfico del polinomio:
    plt.plot(x, y, label=f"Polinomio: {escribir_polinomio(pol)}")

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Gráfico del polinomio")


    # Lo mostramos: 
    plt.grid()
    plt.legend()
    plt.show()
```

```python
graficar_cuadratica([1, 1, 1])
```

```python
# Generar un gráfico de dispersión teniendo en cuenta los siguientes datos [...] se pide que el gráfico tenga como titulo "Relación años/satisfacción"
# A su vez al gráfico agregarle la recta: y = 3.14 + 0.47x, evaluar que pasa.

import matplotlib.pyplot as plt

import numpy as np
```

```python
# Cargo datos:
anios = [8, 7, 10, 3, 6, 13, 4]
satisfaccion = [7, 5, 8, 5, 9, 9, 3]

# Asigno eje X e Y:
plt.scatter(x=anios, y=satisfaccion)

# Personalizo estilos
plt.xlabel("Años")
plt.ylabel("Nivel de satisfacción")
plt.title("Rlación años/satisfacción")

# La recta no es muy viable, debe haber otra manera, se lo tire a chatgtp:
x_values = np.linspace(min(anios), max(anios), 100)  # Generar valores de x
y_values = 3.14 + 0.47 * x_values  # Usar la ecuación de la recta
plt.plot(x_values, y_values, color='red', label='y = 3.14 + 0.47x')


# Muestro el gráfico
plt.legend()
plt.show()
```

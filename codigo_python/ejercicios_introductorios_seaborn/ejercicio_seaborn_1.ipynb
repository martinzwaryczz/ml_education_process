```python
# Cable aclarar que el set de datos de pinguinos viene con seaborn, me rompí la cabeza buscando un csv y encima no andaba :)
# Scatter = dispersión, en este ejercicio se usarán gráficos de dispersión

import seaborn as sns
import matplotlib.pyplot as plt

pinguinos = sns.load_dataset("penguins")
pinguinos.columns = ["especie", "isla", "longitud_pico", "profundidad_pico", "longitud_aleta", "masa", "sexo"]
```

```python
sns.set_theme(style="darkgrid", context="talk")
relacional = sns.relplot(data=pinguinos,
                         height=7,
                         aspect=2,
                         markers=["s", ".", "P", "d"])
relacional.set(title="Análisis de pinguino")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Número de pinguino")
plt.show()
```

```python
# Diagrama de dispersión simple:
# Podemos ver una relación positiva directa entre la longitud del pico y de la aleta
# Aunque estos son ejercicios son introducctorios, no esta demás ir metiendo temas de probabilidad y estadistica

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta")
```

```python
# Teniendo esta relación establecida podemos también dejar evidenciado mediante otro atributo la cantidad de otra caracteristica que se 
# presenta con hue=, pudiendo observar que mientras más aleta y pico más masa, representando 3 datos en un gráfico 2D.
sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="masa")
```

```python
# Podemos también con el hue= asignar una variable cualitativa, haciendo que estas nos den a su vez otra información sobre el comportamiento 
# de nuestros datos

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie")
```

```python
# Si queremos gráficar por un atributo en particular, por ejemplo, el sexo, podemos usar row=
# Esto es aplicable para todos los atributos 

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            row="sexo")
```

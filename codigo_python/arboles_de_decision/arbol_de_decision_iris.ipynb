```python
# Importo dataset a usar

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# Importo librerias

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
# Llevo el dataset a un dataFrame y lo organizo a gusto

iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris_df['Etiqueta'] = iris.target
iris_df.columns = ["Longitud del cepalo", "Ancho del sepalo", "Longitud del petalo", "Ancho del petalo", "Etiqueta"]
```

```python
# Separo los datos

datos_entrenamiento, datos_prueba, clase_entrenamiento, clase_prueba = train_test_split(
    iris_df[["Longitud del cepalo", "Ancho del sepalo", "Longitud del petalo", "Ancho del petalo"]],
    iris_df["Etiqueta"],
    test_size=0.30
)
```

```python
# Creo el arbol de desición

arbol_desicion = tree.DecisionTreeClassifier(criterion="entropy")

arbol = arbol_desicion.fit(datos_entrenamiento, clase_entrenamiento)
```

```python
# score

accuracy = arbol_desicion.score(datos_prueba, clase_prueba)
print(f"La eficiencia es de {round(accuracy, 2)}")
```

```python
# Lo grafico

plt.figure(figsize=(25, 15))
tree.plot_tree(arbol,
               feature_names=["Longitud del sepalo", "Ancho del sepalo", "Longitud del petalo", "Ancho del petalo"],
               class_names=iris.target_names,
               filled=True,
               rounded=True,
               fontsize=12,
               precision=2,
               proportion=True)

plt.title("Árbol de Decisión - Clasificación de Iris", fontsize=16)

plt.show()
```

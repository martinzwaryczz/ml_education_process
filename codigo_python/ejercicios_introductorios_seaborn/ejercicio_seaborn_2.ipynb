```python
# Con el mismo set de datos realizar el mismo gr√°fico con un lineplot 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pinguinos = sns.load_dataset("penguins")
pinguinos.columns = ["especie", "isla", "longitud_pico", "profundidad_pico", "longitud_aleta", "masa", "sexo"]
```

```python
sns.relplot(data=pinguinos,
            x="especie",
            y="masa",
            kind="line", # Ac√° asignamos que sean lineas y no puntos
            hue="sexo",
            style="sexo")

plt.show()
```

```python
# Mismo tipo de grafico, diferentes parametros

sns.relplot(data=pinguinos,
            x="profundidad_pico",
            y="longitud_aleta",
            kind="line",
            hue="especie",
            style="especie")

plt.show()
```

```python
# Scatterplot
sns.relplot(data=pinguinos,
            x="profundidad_pico",
            y="longitud_aleta",
            kind="scatter",
            hue="especie",
            style="especie")

plt.show()
```

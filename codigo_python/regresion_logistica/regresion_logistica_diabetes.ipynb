```python
# Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Levantamos datos
file_patch = "/content/drive/MyDrive/archivos_ml_dl_ds/diabetes2.csv"
df = pd.read_csv(file_patch)
```

```python
# Organizamos datos
x_train = np.array(df["BMI"]).reshape((-1,1))
y_train = np.array(df["Outcome"])
```

```python
# Creamos y entrenamos el modelo

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Predicciones
x_range = np.linspace(df["BMI"].min(), df["BMI"].max(), 300).reshape(-1, 1)
y_pred = model.predict_proba(x_range)[:, 1]
```

```python
# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(x=df["BMI"], y=df["Outcome"], alpha=1)
plt.plot(x_range, y_pred, color='red', label='Probabilidad')
plt.title('Regresión Logística: BMI vs Outcome')
plt.xlabel('BMI')
plt.ylabel('Probabilidad de Diabetes')
plt.legend()
plt.grid()
plt.show()
```

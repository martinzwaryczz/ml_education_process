```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", context="talk", palette="tab10")

consumos = sns.load_dataset("tips")

consumos.columns = ["Total", "Propina", "Genero", "Fumador", "Dia", "Tipo", "Comensales"]
```

# Dispersión
# Stripplot
# Swarmplot
Debemos cambiar en el kind el tipo de gráfico según la categoria de gráficos que querramos. 

```python
color_map = {
    'Thu': 'orange',
    'Fri': 'red',
    'Sat': 'green',
    'Sun': 'blue'
}
```

```python
sns.catplot(x="Dia", y="Total", data=consumos, kind="strip", hue="Dia")
plt.show()
```

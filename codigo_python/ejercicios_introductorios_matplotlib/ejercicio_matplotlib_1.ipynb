```python
# Gr√°ficar sen(x), cos(x) y sen(x)^2+cos(x)^2
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y= np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = (np.cos(x) ** 2) + (np.sin(x) **2)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

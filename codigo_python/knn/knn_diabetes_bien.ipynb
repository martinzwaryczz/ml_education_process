# Ejemplo K vecinos más cercanos

Se observa a continuación la relación entre el IMC de las personas y su edad, además de si tienen o no diabetes.


```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Levanto datos

file_path = '/content/drive/MyDrive/archivos_ml_dl_ds/diabates_knn.csv'
data = pd.read_csv(file_path)
data
```

**Separamos diabeticos de no diabeticos**

```python
diabeticos = data[data["diabetes"] == 1]
no_diabeticos = data[data["diabetes"] == 0]
```

```python
plt.scatter(diabeticos["edad"], diabeticos["imc"], color="red", label="diabeticos")
plt.xlabel("edad")
plt.ylabel("IMC")
plt.title("Relación entre edad e IMC de diabeticos")
plt.legend()
plt.show()
```

```python
plt.scatter(no_diabeticos["edad"], no_diabeticos["imc"], color="green", label="no diabeticos")
plt.xlabel("edad")
plt.ylabel("IMC")
plt.title("Relación entre edad e IMC de no diabeticos")
plt.legend()
plt.show()
```

# Conclusiones
Podemos ver que, independientemente de la edad, mayormente los que mantienen un buen IMC no son diabeticos, caso contrario pasa con los que no, sacando algunos que quizas tengan diabetes por un problema genetico o hereditario.

```python
plt.scatter(diabeticos["edad"], diabeticos["imc"], color="red", label="diabeticos")
plt.scatter(no_diabeticos["edad"], no_diabeticos["imc"], color="green", label="no diabeticos")
plt.xlabel("edad")
plt.ylabel("IMC")
plt.title("Relación entre edad e IMC de diabeticos o no")


plt.legend()
plt.show()
```

```python
datos = data[["edad", "imc"]]
clase = data["diabetes"]
```

```python
clasificador = KNeighborsClassifier(n_neighbors=3)

clasificador.fit(datos, clase)
```

# Ya tenemos el modelo creado
Ya con el modelo creado podrémos clasificar nuevos datos.

En este caso están bien separados las personas con diabetes de las que no, aunque eso no significa que siendo jovenes no puedan tener un gran IMC y diabetes, faltaria completar un poco el dataset pero la idea esta.

```python
edad_paciente_no_diabetico = 22
imc_paciente_no_diabetico = 22

edad_paciente_diabetico = 45
imc_paciente_diabetico = 32

datos_paciente_no_diabetico = [[edad_paciente_no_diabetico, imc_paciente_no_diabetico]]
clasificador.predict(datos_paciente_no_diabetico)

datos_paciente_diabetico = [[edad_paciente_diabetico, imc_paciente_diabetico]]
clasificador.predict(datos_paciente_diabetico)

# Gráficamos igual que antes, pero agregamos la predicción de nuestro modelo de ML:

plt.scatter(diabeticos["edad"], diabeticos["imc"], color="red", label="diabeticos")
plt.scatter(no_diabeticos["edad"], no_diabeticos["imc"], color="green", label="no diabeticos")
plt.scatter(edad_paciente_no_diabetico, imc_paciente_no_diabetico, marker="P", s=50, color="black", label="Paciente no diabetico")
plt.scatter(edad_paciente_diabetico, imc_paciente_diabetico, marker="*", s=50, color="black", label="Paciente diabetico")
plt.xlabel("edad")
plt.ylabel("IMC")
plt.title("Relación entre edad e IMC de diabeticos o no")
plt.legend()


plt.text(0.5, -0.15, f"El paciente diabetico tiene {edad_paciente_diabetico} años y un IMC es {imc_paciente_diabetico}, nuestro modelo lo clasifica como {clasificador.predict(datos_paciente_diabetico)}", ha='center', va='top', transform=plt.gca().transAxes)
plt.text(0.5, -0.20, f"El paciente no diabetico tiene {edad_paciente_no_diabetico} años y un IMC es {imc_paciente_no_diabetico}, nuestro modelo lo clasifica como {clasificador.predict(datos_paciente_no_diabetico)}", ha='center', va='top', transform=plt.gca().transAxes)

plt.show()
```

```python
print(f"Paciente diabetico: {clasificador.predict(datos_paciente_diabetico)} y su probabilidad por clase es de {clasificador.predict_proba(datos_paciente_diabetico)}")
print(f"Paciente no diabetico: {clasificador.predict(datos_paciente_no_diabetico)} y su probabilidad por clase es de {clasificador.predict_proba(datos_paciente_no_diabetico)}")
```

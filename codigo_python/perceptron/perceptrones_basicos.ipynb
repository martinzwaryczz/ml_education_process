# Perceptrones sin librerias

El *perceptron* es el modelo de neurona más básico capaz de formar redes neuronales.

El mismo fue creado por Frank Rosenblatt en 1958, [paper oficial](https://www.ling.upenn.edu/courses/cogs501/Rosenblatt1958.pdf).

Este es eficiente solo en problemas capaz de separarse linealmente, es por esto que si bien fue un invento genial no podía resolver compuertas XOR, aunque Frank dejo bien claro que este fue el primer gran paso para realizar sistemas de aprendizaje más complejos, cabe aclarar que el era neurologo y psicologo.



**Funcionamiento del perceptron**

El perceptron tiene como estructura datos de entrada a los que llamarémos *pesos* y una función de activación que nos dará como resultado 0 o 1, siendo estas la *salida*.

**Perceptrones básicos**

Los perceptrones más básicos son los correspondientes a las compuertas lógicas dado que la función de activacición no es la gran cosa y las entradas tampoco: tendrá pocos pesos, una función de activación básica y un dataset de entrenamiento muy simple.

El pequeño fundamento matemático detrás de estos perceptrones es el siguiente, puede parecer algo tedioso pero con hacerlo uno mismo alcanzará para entender lo realizado:


image.png

```python
import numpy as np

def funcion_de_activacion(x):
        return 1 if x >= 0 else 0
```

```python
pesos_and = np.array([1, 1])
b_and = -2 #
```

**Perceptron AND con casos de prueba**

```python
def perceptron_and(entradas):
  salida_and = np.dot(entradas, pesos_and) + b_and # w1 * x1 + w2 * x2 + wn * xn + b
  return funcion_de_activacion(salida_and)
```

```python
print("Compuerta AND: ")
print("0 AND 0 =", perceptron_and(np.array([0, 0])))  #  0
print("0 AND 1 =", perceptron_and(np.array([0, 1])))  #  0
print("1 AND 0 =", perceptron_and(np.array([1, 0])))  #  0
print("1 AND 1 =", perceptron_and(np.array([1, 1])))  #  1
```

**Perceptron OR con casos de prueba**

```python
pesos_or = np.array([1, 1])
b_or = -0.5
```

```python
def perceptron_or(entradas):
  salida_or = np.dot(entradas, pesos_or) + b_or # w1 * x1 + w2 * x2 + wn * xn + b
  return funcion_de_activacion(salida_or)
```

```python
print("Compuerta OR")
print("0 OR 0 =", perceptron_or(np.array([0, 0])))  #  0
print("0 OR 1 =", perceptron_or(np.array([0, 1])))  #  1
print("1 OR 0 =", perceptron_or(np.array([1, 0])))  #  1
print("1 OR 1 =", perceptron_or(np.array([1, 1])))  #  1
```

**Perceptron NOT con casos de prueba**

```python
peso_not = np.array([-1])
sesgo_not = 0.5
```

```python
def perceptron_not(entradas):
    salida_not = np.dot(entradas, peso_not) + sesgo_not
    return funcion_de_activacion(salida_not)
```

```python
print("Compuerta NOT")
print("NOT 0 =", perceptron_not(np.array([0])))  # Salida debe ser 1
print("NOT 1 =", perceptron_not(np.array([1])))  # Salida debe ser 0
```

**Perceptron NAND con casos de prueba**

```python
pesos_nand = np.array([-2, -2])
sesgo_nand = 3
```

```python
def perceptron_nand(entradas):
  salida_nand = np.dot(entradas, pesos_nand) + sesgo_nand
  return funcion_de_activacion(salida_nand)
```

```python
print("Compuerta NAND")
print("0 NAND 0 =", perceptron_nand(np.array([0, 0])))  # 1
print("0 NAND 1 =", perceptron_nand(np.array([0, 1])))  # 1
print("1 NAND 0 =", perceptron_nand(np.array([1, 0])))  # 1
print("1 NAND 1 =", perceptron_nand(np.array([1, 1])))  # 0
```

**Compuerta NOR** con casos de prueba

```python
pesos_nor = np.array([-1, -1])
sesgo_nor = 0.5
```

```python
def perceptron_nor(entradas):
  salida_nor = np.dot(entradas, pesos_nor) + sesgo_nor
  return funcion_de_activacion(salida_nor)
```

```python
print("Compuerta NOR")
print("0 NOR 0 =", perceptron_nor(np.array([0, 0])))  # 1
print("0 NOR 1 =", perceptron_nor(np.array([0, 1])))  # 0
print("1 NOR 0 =", perceptron_nor(np.array([1, 0])))  # 0
print("1 NOR 1 =", perceptron_nor(np.array([1, 1])))  # 0
```

# Perceptrones con sklearn

Con sklearn el trabajo se simplifica de una manera ENORME.

No será necesario escribir más que unas líneas de código: esto nos permitirá optimizar todo el código anteriormente realizado con cada una de las compuertas.

Esto es super útil, pero nunca esta demás entender el fundamento matemático detrás de las ideas que hoy en día las librerias nos simplifican, aunque me refiero acá como "fundamento matemático" a la idea original, ya que en realidad se utiliza el descenso del gradiente, que es un tema que en este laboratorio no quiero tocar.

```python
from sklearn.linear_model import Perceptron
```

**Perceptron AND con casos de prueba**

```python
# Datos de entrada en X_and, y_and es la salida
X_and = ([[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]])

y_and = np.array([0, 0, 0, 1])
```

```python
perceptron_and = Perceptron()
perceptron_and.fit(X_and, y_and)
```

```python
print("Compuerta AND")
print("0 AND 0 =", perceptron_and.predict([[0, 0]]))  # 0
print("0 AND 1 =", perceptron_and.predict([[0, 1]]))  # 0
print("1 AND 0 =", perceptron_and.predict([[1, 0]]))  # 0
print("1 AND 1 =", perceptron_and.predict([[1, 1]]))  # 1
```

**Perceptron OR con casos de prueba**

```python
X_or = ([[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]])

y_or = np.array([0, 1, 1, 1])
```

```python
perceptron_or = Perceptron()
perceptron_or.fit(X_or, y_or)
```

```python
print("Compuerta OR")
print("0 OR 0 =", perceptron_or.predict([[0, 0]]))  # 0
print("0 OR 1 =", perceptron_or.predict([[0, 1]]))  # 1
print("1 OR 0 =", perceptron_or.predict([[1, 0]]))  # 1
print("1 OR 1 =", perceptron_or.predict([[1, 1]]))  # 1
```

**Perceptron NOT con casos de prueba**

```python
X_not = ([[0],
          [1]])

y_not = ([1, 0])
```

```python
perceptron_not = Perceptron()
perceptron_not.fit(X_not, y_not)
```

```python
print("Compuerta NOT")
print("0 = ", perceptron_not.predict([[0]]))  # 0
print("1 = ", perceptron_not.predict([[1]]))  # 1
```

**Perceptron NAND con casos de prueba**

```python
X_nand = ([[0, 0],
           [0, 1],
           [1, 0],
           [1, 1]])

y_nand = ([1, 1, 1, 0])
```

```python
perceptron_nand = Perceptron(max_iter=10, tol=1e-3, random_state=42) # Ver más sobre esto, inicialmente no anduvo bien
perceptron_nand.fit(X_nand, y_nand)
```

```python
print("Compuerta NAND")
print("0 NAND 0 =", perceptron_nand.predict([[0, 0]]))  # 1
print("0 NAND 1 =", perceptron_nand.predict([[0, 1]]))  # 1
print("1 NAND 0 =", perceptron_nand.predict([[1, 0]]))  # 1
print("1 NAND 1 =", perceptron_nand.predict([[1, 1]]))  # 0
```

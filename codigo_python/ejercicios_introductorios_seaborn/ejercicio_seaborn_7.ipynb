```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vuelos = sns.load_dataset("flights")
vuelos.columns = ["AÃ±o", "Mes", "Cant_Pasajeros"]
```

```python
# Histograma, seria interesante levantar un csv o excel con 2 o 3 datos relacionados para realizar uno destacando las caracteristicas de
# la relaciÃ³n o no, por ej. dia, cantidad de medicinas consumidas por paciente, edad del paciente como hue=
sns.histplot(x="Cant_Pasajeros", data=vuelos)
plt.show()
```

```python
# Este grÃ¡fico es muy analogo al histograma en su uso, solo que al tener una curva constante representa de una mejor manera
# los cambios segÃºn los valores de las variables principalmente cuando estos son muy bruscos.
# Puede graficar distribuciÃ³n bivariada, algo que de momento no vi pero es como las conicas: las vemos en 2d y asÃ­ las grÃ¡ficamos pero
# en realidad es un cono cortado con un plano
sns.kdeplot(x="Cant_Pasajeros", data=vuelos)
plt.show()
```

```python
# Un pequeÃ±o extra es que los grÃ¡ficos pueden tomar tantas maneras de estilos como nos imaginemos, en este caso al usar
# una gran escala de colores ni parece un kdeplot, pero es uno :D
geyser = sns.load_dataset("geyser")

sns.kdeplot(
    data=geyser, x="waiting", y="duration",
    fill=True, thresh=0, levels=100, cmap="mako",
)
```

```python
# Representa la proporciÃ³n o recuento de observaciones que caen por debajo de cada valor Ãºnico en un conjunto de datos
# No me gusto, pero bueno si existe tendrÃ¡ algÃºn uso :P
sns.ecdfplot(x="Cant_Pasajeros", data=vuelos)
plt.show()
```

```python
# Podemos decir que rugplot combina grÃ¡ficos, no sÃ© que mÃ¡s decir, no lo entiendo bien
sns.rugplot(x="Mes", y="Cant_Pasajeros", data=vuelos)
plt.show()
```

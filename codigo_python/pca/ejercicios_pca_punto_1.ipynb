```python
from sklearn.datasets import load_wine # dataset
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
data_vinos = load_wine()
x = data_vinos.data
y = data_vinos.target
```

```python
print(type(data_vinos))
```

```python
pca = PCA(n_components=3) # Queremos que sea un espacio tridimensional, si no ponemos nada será 1.
x_pca = pca.fit_transform(x)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

scatter = ax.scatter(x_pca[:,0], x_pca[:,1], x_pca[:,2], c=y, cmap="viridis") # Se asigna scatter = dado que luego se usará en la leyenda
ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')
ax.set_zlabel('Componente 3')
plt.title('PCA de Datos de Vino en 3D')

legend1 = ax.legend(*scatter.legend_elements(), title="Clases")
ax.add_artist(legend1)

plt.show()
```
